```python
def mysum(N):
    if N<0:
        return 1
    elif N==0:
        return 0
    else:
        return N+mysum(N-1)
```

```python
def mylist(list):
    strlist = []
    intlist = []

    for a in list:
        if type(a) == int or type(a) == float: # checks for type float or int
            intlist.append(a)
        elif type(a) == str: # checks for type str
            strlist.append(a)
    strlist.sort()  # sorts string list
    for i in range(0, len(intlist)):  # sorts the int list
        for j in range(i+1, len(intlist)):
            if intlist[i] >= intlist[j]: # Checks which value is larger
                intlist[i], intlist[j] = intlist[j], intlist[i] # Swaps the index's of the two items if item 1 is larger than item 2

    return intlist+strlist # returns the added list
```

```python
def math(list):
    total = list[0]
    for i in range(1,len(list)-1, 2): # starting at the second item and skipping every other item in the list until the second to last item, this infers that there are just one operator every other number
        operator = list[i] # Separates the operators and the operands
        operand = list[i+1]

        if operator == '+':
            total += operand
        elif operator == '-':
            total -= operand
        elif operator == '*':
            total *= operand
        elif operator == '/':
            if operand == 0:
                print('division by zero is not allowed')
            total /= operand
        elif operator == '^':
            total **= operand
        else:
            print('operator is not recognized, please try again')
    return total

```

```python
mathtest = [2, '^',3]
list1d=[2,'^',2,'+',4,'-',5,'*',10]
print(math(mathtest))
print(math(list1d))
```

```python
from operator import add, sub, mul, pow, truediv

def pemdas(list):
    precedence = {'+':1, '-':1, '*':2, '/':2, '^':3} # order in which operations must be done
    operations = {'+':add, '-':sub, '*':mul, '/':truediv, '^':pow} # each operator's operation

    def tokens(list): #separates the different items in the list into tokens
        tokens = []
        for item in list:
            if isinstance(item, (int, float)) or item in operations:
                tokens.append(item)
            else:
                print('item is not recognized, please try again')
        return tokens

    def apply(operators, values): # applies an operator to the top two elements in the value stack
        operator = operators.pop() # .pop() extracts the operators from the list
        b = values.pop() # extracts the value after the operator
        a = values.pop() # extracts the value before the operator
        values.append(operations[operator](a, b)) # appends the new operation in values

    def eval(tokens): # evaluates using 2 stacks, one for the numbers and one for the operators
        values = []
        operators = []

        for token in tokens:
            if isinstance(token, (int, float)):
                values.append(token) # adds the values that are not operators to the values list
            elif token in operations:
                while (operators and precedence[operators[-1]] >= precedence[token]): # checks the precedence of the operators
                    apply(operators, values)
                operators.append(token)

        while operators: # applies the remaining operators
            apply(operators, values)

        return values[0] # returns the first value in the list

    tokens = tokens(list)
    return eval(tokens) # processes the tokens and evaluates

```

```python
list1d=[2,'^',2,'+',4,'-',5,'*',10]
list2=[10,'*',3,'^',2,'-',6,'+',4]

print(pemdas(list1d))
print(pemdas(list2))
```

```python
import numpy as np
def sin(t,n=1,omega=10,phi=np.pi/2):
    if n==0:
        raise ValueError('n cannot be zero')

    result = np.sin((omega*t)**n+phi)
    return result
```

```python
def quadratic(a,b,c):
    a = np.array(a, dtype=complex)
    b = np.array(b, dtype=complex)
    c = np.array(c, dtype=complex)

    result1 = (-b+np.sqrt(b**2-4*a*c))/(2*a)
    result2 = (-b-np.sqrt(b**2-4*a*c))/(2*a)
    return result1, result2

```

```python
a = np.array([1,2])
b = np.array([2,3])
c = np.array([1,-1])
quadratic(a,b,c)
```
