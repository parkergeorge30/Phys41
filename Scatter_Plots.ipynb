# Scatter Plots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randn(10**3)
y = np.random.randn(10**3)
```

```python
fig,ax = plt.subplots(figsize=(20,8),ncols=2)
ax[0].scatter(x,y)
ax[1].scatter(x,y,s=100,c='orange')
plt.show()
```

```python
sizes = 10*np.random.randn(10**3)**2
colors = np.random.rand(10**3)

fig1,ax = plt.subplots(figsize=(20,8),ncols=3)
ax[0].scatter(x,y)
ax[1].scatter(x,y,s=sizes)
ax[2].scatter(x,y,c=colors,s=sizes, cmap='cool')
plt.show()
```
